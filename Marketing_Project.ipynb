1. Customer Segmentation for Targeted Marketing


```python
# Calculating Frequency as the sum of purchase-related columns
df['Frequency'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# Calculating Monetary as the sum of all monetary columns
df['Monetary'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Now, df contains Recency, Frequency, and Monetary
rfm = df[['Recency', 'Frequency', 'Monetary']]
```

```python
# Function to assign segments based on RFM scores
def rfm_segment(r_score, f_score, m_score):
    if r_score >= 4 and f_score >= 4 and m_score >= 4:
        return 'Champions'
    elif r_score >= 2 and r_score < 4 and f_score >= 3 and f_score < 5 and m_score >= 4:
        return 'Loyal Customers'
    elif r_score >= 3 and f_score >= 2 and m_score >= 2:
        return 'Potential Loyalists'
    elif r_score >= 4 and f_score < 2 and m_score < 2 or r_score >=4:
        return 'New Customers'
    elif r_score >= 3 and r_score < 4 and f_score < 2 and m_score < 2 or f_score >4 and r_score >3 and m_score >4 :
        return 'Promising'
    elif r_score >= 3 and r_score < 4 and f_score >= 3 and m_score >= 3:
        return 'Need Attention'
    elif r_score >= 2 and r_score < 3 and f_score < 3 and m_score < 3:
        return 'About to Sleep'
    elif r_score < 3 and f_score >= 2 and f_score < 5 and m_score >= 2 and m_score < 5:
        return 'At Risk'
    elif r_score < 2 and f_score >= 4 and m_score >= 4:
        return 'Cant Lose Them'
    elif r_score >= 2 and r_score < 3 and f_score >= 2 and f_score < 3 and m_score >= 2 and m_score < 3:
        return 'Hibernating'
    else:
        return 'Lost'
    
```

```python
# Assuming you have the Recency, Frequency, and Monetary values calculated

    
# Define quantiles for RFM metrics
rfm = df[['Recency', 'Frequency', 'Monetary']]

rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])  # Higher score for more recent
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 5, labels=[1, 2, 3, 4, 5])  # Higher score for more frequent
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])  # Higher score for more monetary

# Step 2: Apply the segmentation function
rfm['Segment'] = rfm.apply(lambda x: rfm_segment(x['R_Score'], x['F_Score'], x['M_Score']), axis=1)

# Step 3: Visualize the Segments
print(rfm['Segment'].value_counts())

# Visualize the results
plt.figure(figsize=(10, 6))
sns.countplot(data=rfm, x='Segment', order=rfm['Segment'].value_counts().index, palette='viridis')
plt.title('Number of Customers in Each Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grouping by Segment and calculating mean income
income_analysis = df.groupby(rfm['Segment'])['Income'].mean().reset_index()

# Renaming columns for clarity
income_analysis.columns = ['Segment', 'Average Income']
income_analysis
```

```python
# Plotting the bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x='Segment', y='Average Income', data=income_analysis, palette="viridis")

# Overlaying the line plot
plt.plot(income_analysis['Segment'], income_analysis['Average Income'], marker='D', color='black', linestyle='-', linewidth=2)

# Annotating the points with the exact income values
for index, value in enumerate(income_analysis['Average Income']):
    plt.text(index, value + 0.02, round(value, 2), color='black', ha='center')

# Adding labels and title
plt.title('Average Income by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Average Income')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
#Income Distribution by Customer Segment with Min to Max Income

plt.figure(figsize=(12, 8))
sns.boxplot(x=rfm['Segment'], y='Income', data=df, palette="Set3")
plt.title('Income Distribution by Customer Segment with Min to Max Income')
plt.xticks(rotation=45)
plt.show()
```

Sales Performance Analysis


```python
# Analyze sales performance across different product categories (e.g., wines, fruits).
```

```python
category_df = df[['MntFishProducts', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntSweetProducts', 'MntGoldProds']].sum()
category_df
```

```python
# Creating the bar plot with more appealing design
plt.figure(figsize=(10, 6))
sns.barplot(x=category_df.index, y=category_df.values, palette='viridis')

plt.plot(category_df.index, category_df.values, color='red', marker='D')

for index, value in enumerate(category_df):
    plt.text(index, value + 0.02, round(value, 2), color='black', ha='center')

plt.title('Total Spend by Product Category', fontsize=16, weight='bold')
plt.xlabel('Product Categories', fontsize=14)
plt.ylabel('Total Spend', fontsize=14)


# Displaying the plot
plt.show()
```

```python
# Assuming you have the RFM segments already in your DataFrame
rfm_sales = df.groupby(rfm['Segment'])[['MntFishProducts', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntSweetProducts', 'MntGoldProds',]].sum().reset_index()
rfm_sales
```

```python
# Plotting the stacked bar plot
plt.figure(figsize=(14, 8))

# Loop through each product category and stack the bars
categories = ['MntFishProducts', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntSweetProducts', 'MntGoldProds']

for i, category in enumerate(categories):
    plt.bar(rfm_sales['Segment'], rfm_sales[category], label=category, bottom=rfm_sales[categories[:i]].sum(axis=1))

# Adding titles and labels
plt.title('Sales by Product Category and RFM Segment', fontsize=18, weight='bold')
plt.xlabel('RFM Segments', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)

plt.legend(title='Product Categories')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Create a pivot table for heatmap
pivot_table = rfm_sales.set_index('Segment')

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap='coolwarm', linewidths=.5)

# Adding titles and labels
plt.title('Sales by Product Category and RFM Segment', fontsize=18, weight='bold')
plt.xlabel('Product Categories', fontsize=14)
plt.ylabel('RFM Segments', fontsize=14)

# Display the heatmap
plt.show()
```

```python
# Calculate the acceptance counts for each campaign
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
acceptance_counts = df[campaign_columns].sum()

# Create a DataFrame for easy plotting
acceptance_counts_df = acceptance_counts.reset_index()
acceptance_counts_df.columns = ['Campaign', 'Acceptance Count']
```

```python
# Plotting the acceptance counts with a bar plot and a line plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Campaign', y='Acceptance Count', data=acceptance_counts_df, palette='Blues_d')

# Adding a line plot to the bar plot
plt.plot(acceptance_counts_df['Campaign'], acceptance_counts_df['Acceptance Count'], color='red', marker='o', linestyle='dashdot')

for index, value in enumerate(acceptance_counts):
    plt.text(index, value + 0.2, round(value,2), color='black', ha='center')

# Adding titles and labels
plt.title('Acceptance Counts of Different Campaigns', fontsize=18, weight='bold')
plt.xlabel('Campaign', fontsize=14)
plt.ylabel('Acceptance Count', fontsize=14)

# Displaying the plot
plt.show()
```

```python
# Assuming df contains the necessary columns including RFM segments and campaign acceptance data

# Summing the acceptance counts by RFM segment and campaign
rfm_campaign_acceptance = df.groupby(rfm['Segment'])[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].sum()

# Resetting the index to make 'Segment' a column
rfm_campaign_acceptance = rfm_campaign_acceptance.reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(15, 10))

# Creating the grouped bar plot
rfm_campaign_acceptance.plot(x='Segment', kind='bar', stacked=False, colormap='viridis', width=0.5)

# Adding titles and labels
plt.title('Campaign Acceptance by RFM Segment', fontsize=18, weight='bold')
plt.xlabel('RFM Segment', fontsize=14)
plt.ylabel('Acceptance Count', fontsize=14)

# Adjusting legend
plt.legend(title='Campaign', bbox_to_anchor=(1.05, 1), loc='upper left')

# Displaying the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(12, 8))

# Create a heatmap
sns.heatmap(rfm_campaign_acceptance.set_index('Segment'), annot=True, cmap='YlGnBu', fmt='d', linewidths=.5)

# Add titles and labels
plt.title('Heatmap of Campaign Acceptance by RFM Segment', fontsize=18, weight='bold')
plt.xlabel('Campaigns', fontsize=14)
plt.ylabel('RFM Segment', fontsize=14)

# Display the heatmap
plt.show()
```
